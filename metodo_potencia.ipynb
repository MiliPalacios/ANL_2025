```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Converge al valor propio mas alto. l=x^T A x /(x^T x) definido siempre porque x = 0
#Vector propio derecho : Ax =lx
#          e izquierdo : x^T A = l x^T
def potencia(A,q, tope = 100):
    for k in range(tope):            # PROBLEMA QUE SEA FINITO
        z = A @ q     #Para aproximar el izquierdo: q.T @ A
        q = z / np.linalg.norm(z)   # NORMALIZACION DE z
        nu = q.T @ A @ q
    return nu,q
```
